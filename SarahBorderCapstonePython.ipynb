```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('telco_churn_clean.csv')

print(f"Dataset Shape: {df.shape}")
df.head()

#load and verify
```

```python
churned_customers = df[df['Churn Label'] == 'Yes']

plt.figure(figsize=(10, 6))
sns.histplot(data=churned_customers, x='Tenure in Months',
             bins=72, color='#E63946', kde=True)

plt.title('The First-Year Loyalty Cliff: When Do Customers Churn?',
          fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Customer Tenure (Months)', fontsize=12)
plt.ylabel('Number of Churned Customers', fontsize=12)
plt.xlim(0, 72)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

#The First-Year Loyalty Cliff (Tenure drop-offs)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Internet Type', hue='Churn Label',
              palette=['#2A9D8F', '#E63946'])

plt.title('Churn Breakdown by Internet Service Type', fontsize=14,
          fontweight='bold', pad=15)
plt.xlabel('Internet Type', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.legend(title='Churned?', labels=['No', 'Yes'])
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

#Product Vulnerability (Fiber Optic churn spike)
```

```python
plt.figure(figsize=(10, 6))


sns.boxplot(data=df, x='Churn Label', y='Satisfaction Score', hue='Churn Label', palette=['#2A9D8F', '#E63946'], legend=False)


plt.title('Customer Satisfaction Scores vs. Retention Status', fontsize=14,
          fontweight='bold', pad=15)
plt.xlabel('Did the Customer Churn?', fontsize=12)
plt.ylabel('Satisfaction Score (1-5)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

#Sentiment Analysis (Satisfaction score gaps)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Payment Method', hue='Churn Label', palette=['#2A9D8F', '#E63946'], hue_order=['No', 'Yes'])
plt.title('Churn Risk Breakdown by Payment Method', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Payment Method', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

#Who/How they are paying
```
